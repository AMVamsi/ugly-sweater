```python
def videoToText(video_url): 
  tube = pytube.YouTube(video_url)
  audio = tube.streams.get_audio_only().download()
  model = whisper.load_model("base")
  transcribe = model.transcribe(audio)
  return transcribe['text'] 
 
```

```python
#Removing emojis 
def cleaningText(text): 
  emoji_pattern = re.compile("["
          u"\U0001F600-\U0001F64F"  # emoticons
          u"\U0001F300-\U0001F5FF"  # symbols & pictographs
          u"\U0001F680-\U0001F6FF"  # transport & map symbols
          u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                            "]+", flags=re.UNICODE)
  return emoji_pattern.sub(r'', text) # no emoji
```

```python
import string 
def removePunctuations(text): 
  return text.translate(str.maketrans('', '', string.punctuation))
```

```python
def removeDigits(text): 
  remove_digit_map = dict.fromkeys(map(ord,string.digits))
  return text.translate(remove_digit_map)
```

```python
df['posts'] = df['posts'].apply(lambda x: cleaningText(x))
df['posts'] = [s.lower() for s in df['posts']]
df['posts'] = df['posts'].apply(lambda x: removePunctuations(x))
df['posts'] = df['posts'].apply(lambda x: removeDigits(x))
```
